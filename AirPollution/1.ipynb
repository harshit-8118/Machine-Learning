```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Train/Train.csv')
```

```python
df.head()
```

```python
data = df.values
X = data[:, :-1]
Y = data[:, -1].reshape((-1,1))
print(X, Y)
```

```python
m = X.shape[0]
m
```

```python
ones = np.ones((m, 1))
X = np.hstack((ones, X))
print(X)
```

```python
def hypothesis(X, theta):
    Y_ = np.dot(X, theta)
    return Y_
```

```python
def error(X, Y, theta):
    Y_ = hypothesis(X, theta)
    te = np.sum((Y_ - Y) ** 2)
    return te / m
```

```python
def gradient(X, Y, theta):
    ypred = hypothesis(X, theta)
    grad = np.dot(X.T, (ypred - Y))
    return grad / m
```

```python
def gradientDescent(X, Y, alpha = 0.1, iters = 200):
    n = X.shape[1]
    theta = np.zeros((n, 1))
    el = []
    for i in range(iters):
        grad = gradient(X, Y, theta)
        e = error(X, Y, theta)
        el.append(e)
        theta = theta - alpha * grad
    return theta, el
```

```python
theta, el = gradientDescent(X, Y)
# print(theta)
# print(el)
```

```python
Y_ = hypothesis(X, theta)
print(Y_)
```

```python
num = (Y_ - Y) ** 2
den = (Y - np.mean(Y)) ** 2
rsq = (1 - (np.sum(num) / np.sum(den))) * 100
print(rsq)
```
