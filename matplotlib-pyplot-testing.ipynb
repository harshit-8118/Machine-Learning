```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = x ** 2 + 3
```

```python
print(x, y)
```

```python
plt.style.use("dark_background")
plt.style.available
```

```python
plt.plot(x, y, label="stock price 1")
plt.plot(x, x**2, label="stock price 2")
plt.ylabel("price")
plt.xlabel("timestamp")
plt.title("My Stock Exchange")
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, label="stock price 1")
plt.scatter(x, x**2, label="stock price 2")
plt.ylabel("price")
plt.xlabel("timestamp")
plt.title("My Scatter Stock Exchange")
plt.legend()
plt.show()
```
