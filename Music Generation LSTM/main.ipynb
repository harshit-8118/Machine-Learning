```python
elems_to_parse = midi.flat.notes
```

```python
c = 0
for e in elems_to_parse:
    if not isinstance(e, chord.Chord):  
        c += 1
        print(e, e.offset, e.pitch)
    else:
        for pitch in e.pitches:
            c += 1
            print(e, e.offset, pitch)
print(c)
```

```python
def store_pitches(elems_to_parse):
    notes_demo = []
    for e in elems_to_parse:
        # if e is note, store its pitch
        if e.isNote:
            notes_demo.append(str(e.pitch))
            
        # if e in chord, then split each note of chord ans join with +
        elif isinstance(e, chord.Chord):
            notes_demo.append("+".join([str(p) for p in e.normalOrder]))
            for p in e.pitches:
                notes_demo.append(str(p))
    return notes_demo
```

```python
notes = []
def process_all_songs():
    for i, f in enumerate(glob.glob("midi_songs/*/*")):
        song = converter.parse(f)
        element_to_parse = song.flat.notes
        pitches = store_pitches(element_to_parse)
        notes.extend(pitches)
        print(i, end=" ")
        
process_all_songs()
```

```python
with open('notes.pkl', 'wb') as f:
    pickle.dump(notes, f)

notes = []
with open('notes.pkl', 'rb') as f:
    notes = pickle.load(f)
```
