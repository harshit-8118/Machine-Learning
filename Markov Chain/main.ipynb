```python
import numpy as np
```

```python
text = "the man was ...they...then...the ...the  "

# |   X     |    Y   |   freq
# |   the   |    _   |   4
# |   the   |    y   |   1
# |   the   |    n   |   1
# |   man   |    _   |   1
```

```python
def generateTable(data, k=4):
    T = {}
    for i in range(len(data)-k):
        X = data[i: i + k]
        Y = data[i + k]
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
        elif T[X].get(Y) is None:
            T[X][Y] = 1
        else:
            T[X][Y] += 1

    for kx in T.keys():
        for ki in T[kx].keys():
            s = list(T[kx].values())
            s = np.sum(s)
            T[kx][ki] = T[kx][ki] / s

    return T

T = generateTable(text)
```

```python
text_path = 'speech.txt'
def load_text(text_path):
    with open(text_path, encoding='utf-8') as f:
        return f.read().lower()

text = load_text(text_path)
```

```python
k = 6
def trainMarkovChain(text, k):
    T = generateTable(text, k)
    return T
```

```python
def sample_next(ctx, T, k):
    ctx = ctx[-k:]
    # print(ctx)
    if T.get(ctx) is None:
        return " "
        
    possible_chars = list(T[ctx].keys())
    possible_values = list(T[ctx].values())
    sum_values = np.sum(possible_values)
    normalized_probs = [prob / sum_values for prob in possible_values]
    
    # print(possible_chars)
    # return possible_chars
    return np.random.choice(possible_chars, p=normalized_probs)
```

```python
def Generate_text(start_sentence, T, k, max_len=2000):
    
    sentence = start_sentence
    ctx = start_sentence
    for i in range(max_len):
        p = sample_next(ctx, T, k)
        sentence += p[0]
        ctx = sentence[-k:]

    return sentence
```
