```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dfx = pd.read_csv('Training Data/Diabetes_XTrain.csv')
dfy = pd.read_csv('Training Data/Diabetes_YTrain.csv')

print(dfx.head())
print(dfy.head())
```

```python
X = dfx.values
Y = dfy.values

print(X)
print(Y)
```

```python
dftest = pd.read_csv('Test Cases/Diabetes_Xtest.csv')

# print(dftest.head())
testX = dftest.values
testX = np.int_(testX)
print(testX)
```

```python
def dist(X1, X2):
    return np.sqrt(sum((X1 - X2) ** 2))

def knn(X, Y, query, k = 5):
    vals = []
    m = X.shape[0]
    for i in range(m):
        d = dist(X[i], query)
        vals.append((d, Y[i]))

    vals = sorted(vals)[:k]
    pred = [v[1] for v in vals]
    predictions, counts = np.unique(pred, return_counts=True)
    maxfreq = predictions[np.argmax(counts)]
    return maxfreq
```

```python
ys = []
for i in range(testX.shape[0]):
    v = knn(X, Y, testX[i], 10)
    ys.append(v)

print(ys)
```

```python
df_ys = pd.DataFrame(ys, columns=['Outcome'])
df_ys.to_csv('Outcome.csv', sep=',', index=False)
```
