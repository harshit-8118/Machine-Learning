```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('images/4.jpg')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
print(img.shape)
```

```python
# matplot lib does not requires to convert because it's it own function and well defined init.
cv2.imshow("Image", img)
cv2.waitKey(0) # this basicallly tells cv2 window to wait infinitally 
# cv2.waitKey(250) # this basicallly tells cv2 window to wait 250 milliseconds 
cv2.destroyAllWindows()
```

```python
# matplot lib requires to convert "RGB2BGR" because matplot understands image reverse of rgb to bgr and vice versa
plt.imshow(rgb_img)
plt.show()
plt.axis("off")
```

```python
crop_img = rgb_img[220:430, 320:530]
plt.imshow(crop_img)
plt.show()
plt.axis("off")
crop_img.shape
```

```python
crop_img = cv2.resize(crop_img, (100,100))
crop_img.shape
```

```python
plt.imshow(crop_img)
plt.title("Dr Strange" + str(crop_img.shape))
plt.axis("off")
plt.show()
```
