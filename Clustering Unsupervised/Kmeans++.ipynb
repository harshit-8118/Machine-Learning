```python
import matplotlib.pyplot as plt, numpy as np
from sklearn.datasets import make_blobs
```

```python
X, Y, centers = make_blobs(n_samples=500, n_features=2, random_state=3, centers=5, return_centers=True)
```

```python
centers
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.scatter(centers[:,0], centers[:,1], marker='^', color='red')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
cent = kmeans.cluster_centers_
cent
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.scatter(cent[:, 0], cent[:, 1], marker='^', color='red')
plt.scatter(centers[:, 0], centers[:, 1], marker='*', color='black')
```
