```python
train_gen = ImageDataGenerator(rescale=1.0/255)
train_generator = train_gen.flow_from_directory(
    'Images/',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
)

val_gen = ImageDataGenerator(rescale=1.0/255)
val_generator = val_gen.flow_from_directory(
    'val_images/',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
)
```

```python
from keras import models, utils
from keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
model = models.Sequential()
```

```python
model.add(Convolution2D(filters=32, kernel_size=(5,5), padding='same', input_shape=(150, 150, 3), activation='relu'))
model.add(Convolution2D(filters=64, kernel_size=(3,3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Convolution2D(filters=64, kernel_size=(5,5), padding='same', activation='relu'))
model.add(Dropout(0.25))
model.add(Convolution2D(filters=16, kernel_size=(3,3), padding='same', activation='relu'))
model.add(Convolution2D(filters=8, kernel_size=(3,3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(20, activation='relu'))
model.add(Dense(4, activation='softmax'))
model.summary()
```

```python
adam = optimizers.Adam(learning_rate=1e-4)
model.compile(optimizer=adam, loss='categorical_crossentropy', metrics=['accuracy'])
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
hist = model.fit_generator(
    train_generator,
    epochs=10,
    validation_data=val_generator,
)
```

```python
h = hist.history
```

```python
acc = h['accuracy']
val_acc = h['val_accuracy']
loss = h['loss']
val_loss = h['val_loss']
plt.plot(acc, label='training accuracy')
plt.plot(loss, label='training loss')
plt.plot(val_acc, label='validation accuracy')
plt.plot(val_loss, label='validation loss')
plt.legend()
```
