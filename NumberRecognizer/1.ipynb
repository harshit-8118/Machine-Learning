```python
def dist(X, Y):
    return np.sqrt(sum((X - Y) ** 2))
    
```

```python
def knn(X, Y, QP):
    vals = []
    k = 5
    vals = np.sqrt(np.sum((X - QP) ** 2, axis=1)).reshape((X.shape[0],1))
    Y = np.round(Y.reshape((Y.shape[0], 1)))
    vals = np.concatenate((vals, Y), axis=1)
    column_to_sort = vals[:, 0]
    sorted_indices = np.argsort(column_to_sort)
    vals = vals[sorted_indices]
    vals = vals[:k]
    vals = [v[1] for v in vals]
    pred, cnts = np.unique(vals, return_counts=True)
    result = pred[np.argmax(cnts)]
    return result
```
